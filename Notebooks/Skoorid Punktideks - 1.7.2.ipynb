```python
def teisenda_vahemikku(algne, maksimum, miinimum):
    uus = ((algne - miinimum) / (maksimum - miinimum))
    
    if uus > 1:
        return 1.0
    elif uus < 0:
        return 0.0
    else:
        return uus
```

```python
# Algse kausta nimi, võtab sealt algsed failinimed, ei loe neid faile sisse
source = "../etnc19_web_2019_andmed/"
target = "../etnc19_web_2019_skoorid/"

os.makedirs(os.path.dirname(target), exist_ok=True)

# Avab järjest kõik failid algkaustas
for file in [f for f in os.listdir(source)]:
    
    source_filename = os.path.join(source, file)
    with open(source_filename, "r", encoding="UTF-8") as fr:
        andmed = json.load(fr)
    
    skoorid = dict()
    
    formaalsus_ilma, formaalsus_kaalud, formaalsus_kaalud_sujuv = formaalsus(andmed)
    
    skoorid["formaalsus_kaalud_sujuv"] = teisenda_vahemikku(formaalsus_kaalud_sujuv, 176, -226)
    skoorid["formaalsus_kaalud"] = teisenda_vahemikku(formaalsus_kaalud, 366, -448)
    skoorid["formaalsus_ilma"] = teisenda_vahemikku(formaalsus_ilma, 34, -45)
    
    spontaansus_ilma, spontaansus_kaalud, spontaansus_kaalud_sujuv = spontaansus(andmed)
    
    skoorid["spontaansus_kaalud_sujuv"] = teisenda_vahemikku(spontaansus_kaalud_sujuv, 268, -103)
    skoorid["spontaansus_kaalud"] = teisenda_vahemikku(spontaansus_kaalud, 547, -227)
    skoorid["spontaansus_ilma"] = teisenda_vahemikku(spontaansus_ilma, 53, -15)
    
    target_filename = os.path.join(target, file)
    
    with open(target_filename, "w", encoding="UTF-8") as fw:
        json.dump(skoorid, fw, sort_keys=True, indent=4)
```

```python
target = "../minikorpus/skoorid_koos_vahed/"
source = "../minikorpus/skoorid_koos/"

skooristik = defaultdict(lambda: 0)

os.makedirs(os.path.dirname(target), exist_ok=True)

for filename, form, spon in zip(df['failinimi'], df['formaalsus'], df['spontaansus']):
    skoornimi = ".".join(filename.split(".")[:-1]) + "_andmed.json"
    
    source_filename = os.path.join(source, skoornimi)
    with open(source_filename, "r", encoding="UTF-8") as fr:
        skoorid = json.load(fr)
    
    skoorid["formaalsus_ilma"] = abs(skoorid["formaalsus_ilma"] - form)
    skoorid["formaalsus_kaalud"] = abs(skoorid["formaalsus_kaalud"] - form)
    skoorid["formaalsus_kaalud_sujuv"] = abs(skoorid["formaalsus_kaalud_sujuv"] - form)
        
    skoorid["spontaansus_ilma"] = abs(skoorid["spontaansus_ilma"] - spon)
    skoorid["spontaansus_kaalud"] = abs(skoorid["spontaansus_kaalud"] - spon)
    skoorid["spontaansus_kaalud_sujuv"] = abs(skoorid["spontaansus_kaalud_sujuv"] - spon)
    
    skooristik["arv"] += 1
     
    skooristik["formaalsus_ilma"] += skoorid["formaalsus_ilma"]
    skooristik["formaalsus_kaalud"] += skoorid["formaalsus_kaalud"]
    skooristik["formaalsus_kaalud_sujuv"] += skoorid["formaalsus_kaalud_sujuv"]
        
    skooristik["spontaansus_ilma"] += skoorid["spontaansus_ilma"]
    skooristik["spontaansus_kaalud"] += skoorid["spontaansus_kaalud"]
    skooristik["spontaansus_kaalud_sujuv"] += skoorid["spontaansus_kaalud_sujuv"]
    
    with open(target + filename, "w", encoding="UTF-8") as fw:
        json.dump(skoorid, fw, sort_keys=True, indent=4)
```
