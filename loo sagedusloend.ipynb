```python
from collections import defaultdict
import os
import re
from estnltk import Text
import string
from tqdm.notebook import tqdm
import pandas as pd
```

```python
from estnltk.taggers import VabamorfTagger
# Create morphological tagger with default settings
morph_tagger = VabamorfTagger(guess=False, propername=False, disambiguate=False)
```

```python
morph_tagger
```

Valimi sagedusloend

```python
# Algse kausta nimi, võtab sealt algsed failinimed, ei loe neid faile sisse
source = "etnc19_web_2019_100000/"

dictionary = defaultdict(int)

i = 0

# Avab järjest kõik failid algkaustas
for file in [f for f in os.listdir(source)]:

    with open(os.path.join(source, file), "r", encoding="UTF-8") as f:
        # Loeb failist vaid sisu, ignoreerib alguses olevat metainfot
        text = Text("".join(f.readlines()[1:]))
        text.tag_layer(['words', 'sentences', 'compound_tokens'])

        morph_tagger.tag( text )

        text.tag_layer(['morph_analysis'])

        for word in text['morph_analysis']:
            # vaatab, et algustäht poleks suur
            pos = word.partofspeech
            if len(pos) == 1 and pos[0] == None:
                m = word.text
                if not all(j in string.punctuation for j in m):
                    if not (m[0].isupper() and not m.isupper()):
                        dictionary[m.lower()] += 1
```

```python
dictionary
```

```python
sorted_dict = sorted(dictionary.items(), key=lambda kv: -kv[1])
```

```python
with open('tundmatud_esinemised.txt', 'w', encoding="UTF-8") as f:
    for line in sorted_dict:
        f.write(str(line))
        f.write("\n")
```

Minikorpuse sagedusloend

```python
minikorpus_dir = "minikorpus/"
minikorpus = pd.read_csv(minikorpus_dir+'skoorid.csv')

minikorpus_formaalsus = minikorpus[minikorpus["DIMENSIOON"]=="form"][["FAIL", "KESKMINE SKOOR"]].rename(columns={"KESKMINE SKOOR": "Formaalsus_Keskmine"})
minikorpus_spontaansus = minikorpus[minikorpus["DIMENSIOON"]=="form"][["FAIL", "KESKMINE SKOOR"]].rename(columns={"KESKMINE SKOOR": "Spontaansus_Keskmine"})
```

```python
minikorpus_failide_keskmised = minikorpus_formaalsus.copy()
```

```python
minikorpus_failide_keskmised["Spontaansus_Keskmine"] = minikorpus_spontaansus["Spontaansus_Keskmine"]
```

```python
mfk = minikorpus_failide_keskmised.fillna(0)
```

```python
minikorpus_dict = dict(zip(mfk["FAIL"], zip(mfk["Formaalsus_Keskmine"], mfk["Spontaansus_Keskmine"])))
```

```python
minikorpus_dict
```

```python
dictionary = defaultdict(lambda: [0, []])

i = 0


for failinimi, keskmised in minikorpus_dict.items():
    with open(os.path.join("minikorpus\\tekstid", failinimi), "r", encoding="UTF-8") as f:
        # Loeb failist vaid sisu, ignoreerib alguses olevat metainfot
        text = Text("".join(f.readlines()[1:]))
        text.tag_layer(['words', 'sentences', 'compound_tokens'])

        morph_tagger.tag( text )

        text.tag_layer(['morph_analysis'])

        for word in text['morph_analysis']:
            # vaatab, et algustäht poleks suur
            pos = word.partofspeech
            if len(pos) == 1 and pos[0] == None:
                m = word.text
                if not all(j in string.punctuation for j in m):
                    if not (m[0].isupper() and not m.isupper()):
                        sissekanne = dictionary[m.lower()]
                        sissekanne[0] += 1
                        
                        loend = (failinimi, keskmised)
                        
                        if loend not in sissekanne[1]:
                            sissekanne[1].append(loend)
                            
                        dictionary[m.lower()] = sissekanne
```

```python
dictionary
```

```python
sorted_dict = sorted(dictionary.items(), key=lambda kv: -kv[1][0])
```

```python
with open('minikorpuse_tundmatud_esinemised.txt', 'w', encoding="UTF-8") as f:
    for line in sorted_dict:
        f.write(str(line))
        f.write("\n")
```
